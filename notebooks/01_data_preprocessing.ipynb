```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import matplotlib.pyplot as plt

# fetch dataset
heart_disease = fetch_ucirepo(id=45)

# data (as pandas dataframes)
X= pd.DataFrame(heart_disease.data.features ,columns=heart_disease.data.features_names)
y=pd.Series(heart_disease.data.targets.values.ravel(), name="num")

# metadata
print(heart_disease.metadata)

# variable information
print(heart_disease.variables)
```

```python
df = pd.concat([X,y],axis=1)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['ca'] = df['ca'].fillna(df['ca'].mode()[0])
df['thal'] = df['thal'].fillna(df['thal'].mode()[0])
```

```python
non_numeric_cols = df.select_dtypes(exclude=["number"]).columns
print("Non-numeric columns:", non_numeric_cols.tolist())
```

```python
df['age'].plot(kind='hist',bins=10, figsize=(8,6), title= 'Age Distribution')
plt.show()
```

```python
df.plot(kind='box' ,figsize=(10,6))
```

```python
# Standardize numerical features
from sklearn.preprocessing import MinMaxScaler

# Copy the dataframe to keep the original safe
scaled_df = df.copy()

# Features to scale (numerical ones)
num_features = ["age", "trestbps", "chol", "thalach", "oldpeak", "ca"]

# Initialize the scaler
scaler = MinMaxScaler()

# Fit and transform
scaled_df[num_features] = scaler.fit_transform(scaled_df[num_features])
scaled_df.head()
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Heart Disease Features")
plt.show()
```

```python
scaled_df.to_csv("scaled_heart.csv", index=False)

```
